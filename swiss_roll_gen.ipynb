```python
from sklearn.datasets import make_swiss_roll
import numpy as np
import csv
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.svm import SVC
```

```python
X, t= make_swiss_roll(2000, noise = 0.1, random_state=42)
```

```python
t_mean = t.mean()
print(t_mean)
```

```python
classif = []
for el in t:
    if el > t_mean:
        classif.append(1)
    else:
        classif.append(-1)
```

```python
with open('swiss_roll_data.csv', 'w', newline='') as archivo_csv:
        escritor_csv = csv.writer(archivo_csv)
        escritor_csv.writerows(X)
```

```python
with open('swiss_roll_class.csv', 'w', newline='') as archivo_csv:
        escritor_csv = csv.writer(archivo_csv)
        for el in classif:
            escritor_csv.writerow([el])
```
